```python
movies.columns = ['DayOfWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross', 'BudgetMillions', 'GrossMillions',
       'IMDbRating', 'MovieLensRating', 'OverseasMillions', 'Overseas',
       'ProfitMillions', 'Profit', 'RuntimeMinutes', 'USMillions',
       'GrossUS']
```

```python
movies.Genre = movies.Genre.astype('category')
```

```python
FilterStudios = (movies.Studio == 'Sony') | (movies.Studio == 'Buena Vista Studios') | (movies.Studio == 'Universal') | (movies.Studio == 'WB') | (movies.Studio == 'Paramount Pictures') | (movies.Studio == 'Fox')
FilterGenre = (movies.Genre == 'action') | (movies.Genre == 'comedy') | (movies.Genre == 'adventure') | (movies.Genre == 'animation') | (movies.Genre == 'drama')
movies[FilterStudios]

```

```python
studio_filters = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']
genre_filters = ['action', 'adventure', 'animation', 'comedy', 'drama']
```

```python
filtered_movies = movies[movies.Genre.isin(genre_filters)]
```

```python
filtered_movies = filtered_movies[filtered_movies.Studio.isin(studio_filters)]
```

```python
filtered_movies
```

```python
sns.set(style='darkgrid', palette='muted', color_codes = True)
fig, ax = plt.subplots()
fig.set_size_inches(12, 10)
box1 = sns.boxplot(data=filtered_movies, x='Genre', y='GrossUS', orient='v', color='lightgray')
box1 = sns.stripplot(x="Genre", y="GrossUS", data=filtered_movies, hue="Studio", jitter= True, linewidth=0, alpha=0.7,  palette=['Blue', 'Green', 'Red', 'Purple', 'Gold', 'LightBlue'], dodge=False)

ax.legend(prop={'size':20}, loc=2, bbox_to_anchor=(1.05, 1) )
plt.title('Domestic Gross % by Genre', fontsize=35, color='Black', fontname='Console')
plt.ylabel('Gross % US', fontsize=25, color='Black', fontname='Console')
plt.xlabel('Genre', fontsize=25, color='Black', fontname='Console')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
